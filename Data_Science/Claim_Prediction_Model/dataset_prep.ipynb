```python
import pandas as pd

# Load the uploaded CSV files
file1_path = '../Health_Claims/Train_Beneficiarydata-1542865627584.csv'
file2_path = '../Health_Claims/Train_Inpatientdata-1542865627584.csv'
file3_path = '../Health_Claims/Train_Outpatientdata-1542865627584.csv'
file4_path = '../Health_Claims/Train-1542865627584.csv'

beneficiary_data = pd.read_csv(file1_path)
inpatient_data = pd.read_csv(file2_path)
outpatient_data = pd.read_csv(file3_path)
train_data = pd.read_csv(file4_path)

# Display the first few rows of each dataframe to understand their structure
beneficiary_data_head = beneficiary_data.head()
inpatient_data_head = inpatient_data.head()
outpatient_data_head = outpatient_data.head()
train_data_head = train_data.head()

beneficiary_data_head
```

```python
# Merge the beneficiary data with inpatient data
merged_inpatient = pd.merge(inpatient_data, beneficiary_data, on='BeneID', how='left')
merged_inpatient
```

```python
# Merge the beneficiary data with outpatient data
merged_outpatient = pd.merge(outpatient_data, beneficiary_data, on='BeneID', how='left')
merged_outpatient
```

```python
# Concatenate the inpatient and outpatient data
combined_data = pd.concat([merged_inpatient, merged_outpatient], ignore_index=True)
combined_data
```

```python
# Merge the combined data with train data on the 'Provider'
final_merged_data = pd.merge(combined_data, train_data, on='Provider', how='left')

# Display the first few rows of the final merged data
final_merged_data.head()
```

```python
data = final_merged_data.sample(frac= 0.1)
data
```

```python
data.to_csv('health_claims.csv')
```
